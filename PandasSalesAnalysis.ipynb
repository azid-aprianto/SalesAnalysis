**EXPLORATORY DATA ANALYSIS**

- <H4> What is the overall sales trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# Getting month year for the dataset
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# Grouping month year
df_trend = df.groupby('month_year').sum()['sales'].reset_index()
```

```python
# Setting the figure size
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'], df_trend['sales'])
plt.xticks(rotation ='vertical', size=8)
plt.show()
```

- <h4> Which are the Top 10 products by sales?

```python
df.head(3)
```

```python
# Grouping product name column
prod_sales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
# Sorting produk_sales column
prod_sales = prod_sales.sort_values('sales',ascending=False)
```

```python
# Top 10 by product by sales
prod_sales[:10]
```

- <h4> Which are the Most Selling Products?

```python
# Grouping product name
most_sell_prod = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
# Sorting most sell product
most_sell_prod = most_sell_prod.sort_values('quantity', ascending=False)
```

```python
most_sell_prod[:10]
```

- <h4> Which is the most preferred Ship Mode?

```python
# Setting figure size
plt.figure(figsize=(16,8.5))

# Plotting ship mode
sns.countplot(df['ship_mode'])
plt.show()
```

- <h4> Which are the Most Profitable Category and Sub-Category?

```python
# Grouping category and subcategory
cat_subcat_profit = pd.DataFrame(df.groupby(['category', 'sub_category']).sum()['profit'])
```

```python
# Sorting the values
cat_subcat_profit.sort_values(['category', 'profit'], ascending=False)
```
